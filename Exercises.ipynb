```python
import numpy as np
import time
import matplotlib.pyplot as plt
import torch
```

# Exercise 1.1

```python
def compute_time(n=10**8):
    tms=[]
    for j in range(10):
        a=np.random.randn(n)
        b=np.random.randn(n)
        start=time.time()
        c=np.dot(a,b)
        end=time.time()
        tm=end-start
        tms.append(tm)
    return tms

def compute_time_gpu(n=10**8):
    device=torch.device('cuda')
    tms=[]
    for j in range(10):
        a=torch.tensor(np.random.randn(n)).to(device)
        b=torch.tensor(np.random.randn(n)).to(device)
        start=time.time()
        c=torch.dot(a,b)
        end=time.time()
        tm=end-start
        tms.append(tm)
    return tms    
```

```python
tms_gpu=compute_time_gpu()
```

```python
plt.plot(tms_gpu)
```

```python
tms_cpu=compute_time()
```

```python
plt.plot(tms_cpu)
```

```python
#Estimate Gflops/sec
gflops_gpu=(2*(10**8)-1)/(np.mean(tms_gpu)*10**9)
gflops_cpu=(2*(10**8)-1)/(np.mean(tms_cpu)*10**9)
print(gflops_gpu,gflops_cpu)
```
